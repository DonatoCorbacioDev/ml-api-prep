# Analisi non supervisionata - Titanic

Obiettivo: esplorare il dataset senza utilizzare la variabile target per individuare pattern o gruppi di passeggeri simili.

```python
import pandas as pd

df = pd.read_csv("../data/titanic.csv")
```

In questa analisi non viene utilizzata la variabile 'Survived', in quanto l'obiettivo è esplorativo. 

Per l'analisi non supervisionata vengono selezionate solo variabili numeriche e facilmente interpretabili.

Feature consigliate: Age, Fare, Pclass, SibSp, Parch

```python
from sklearn.preprocessing import StandardScaler

features = ["Age", "Fare", "Pclass", "SibSp", "Parch"]
X = df[features].dropna()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

X["Cluster"] = clusters
X.head()
```

```python
df_unsup = df.loc[X.index].copy()
df_unsup["Cluster"] = clusters
df_unsup.head()

df_unsup.groupby("Cluster")[features].mean()
```

Dall'analisi emergono tre cluster distinti:
- un gruppo di passeggeri benestanti che viaggiano spesso in famiglia.
- un gruppo di passeggeri giovani in terza classe con biglietti economici.
- un gruppo di passeggeri più anziani, di classe alta, che viaggiono prevalentemente da soli.

I cluster risultano coerenti e facilmente interpretabili, confermando la presenza di segmenti naturali nel dataset.
